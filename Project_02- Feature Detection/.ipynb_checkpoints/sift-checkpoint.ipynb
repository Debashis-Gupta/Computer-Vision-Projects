```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Read image

```python
img = cv2.imread("./Image/andrew.JPG",0)
plt.imshow(img,cmap='gray')
```

### Rotating the image


```python
def image_rotation(image,angle):
    height,width = image.shape[:2]
    center = (width/2,height/2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle,1.0)
    rotated_image = cv2.warpAffine(image,rotation_matrix,(width,height))
    return rotated_image
rotated_image = image_rotation(img,-30)
plt.imshow(rotated_image,cmap='gray')
```

```python
cv2.imshow("Orginal Image",img)
cv2.imshow("Rotated Image",rotated_image)
cv2.waitKey(0)
```

#### SIFT Detector

```python
sift = cv2.SIFT_create()
img1=img
kp1,descriptor1 = sift.detectAndCompute(img1,None)
kp2,descriptor2 = sift.detectAndCompute(rotated_image,None)

imgKp1 = cv2.drawKeypoints(img1,kp1,None)
imgKp2 = cv2.drawKeypoints(rotated_image,kp2,None)

cv2.imshow("Kp1",imgKp1)
cv2.imshow("Kp2",imgKp2)
cv2.waitKey(0)
```
